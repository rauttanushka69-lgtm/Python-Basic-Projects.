```python
# Q1 to Q2: Load + Explore
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('train.csv')
df.head(10)
```

```python

print("""Reason: I chose House Prices dataset because target 'SalePrice' is continuous.
We can predict house price using features like Area, Quality.
This is useful for real estate.""")
```

```python
missing = pd.DataFrame({'Total': df.isnull().sum(), '%': (df.isnull().sum()/len(df))*100})
print(missing[missing['Total']>0].sort_values('%', ascending=False))

df['LotFrontage'].fillna(df['LotFrontage'].median(), inplace=True)
df.fillna(0, inplace=True)
print(df.duplicated().sum())
df.drop_duplicates(inplace=True)
```

```python
print(df.describe())
print(df['SalePrice'].min())
print(df['SalePrice'].max())
print(df['SalePrice'].mean())
print(df['SalePrice'].median())
```

```python
sns.histplot(df['SalePrice'], kde=True)
plt.show()
sns.histplot(df['GrLivArea'], kde=True)
plt.show()
sns.histplot(df['TotalBsmtSF'], kde=True)
plt.show()
```

```python
sns.countplot(x='OverallQual', data=df)
plt.xticks(rotation=45)
plt.show()
sns.countplot(x='Neighborhood', data=df)
plt.xticks(rotation=45)
plt.show()
sns.countplot(x='HouseStyle', data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), cmap='coolwarm')
plt.show()
```
