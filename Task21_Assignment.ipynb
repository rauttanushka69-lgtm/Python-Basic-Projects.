```python
!pip install gradio pandas scikit-learn -q

import gradio as gr
import pandas as pd
import pickle

# 1. Load your trained model, scaler and columns
model = pickle.load(open('LR_model.pkl', 'rb'))
scaler = pickle.load(open('scaler.pkl', 'rb'))
model_columns = pickle.load(open('columns.pkl', 'rb'))

# 2. Define prediction function
def predict_price(year, km_driven, tax, mpg, engine_size):
    # Create dataframe for prediction
    input_data = pd.DataFrame({
        'year': [year],
        'mileage': [km_driven],
        'tax': [tax],
        'mpg': [mpg],
        'engineSize': [engine_size]
    })

    # Encode categorical if any - match your training columns
    input_encoded = pd.get_dummies(input_data)
    input_encoded = input_encoded.reindex(columns=model_columns, fill_value=0)

    # Scale and Predict
    input_scaled = scaler.transform(input_encoded)
    prediction = model.predict(input_scaled)

    return f"Predicted Price: £ {round(prediction[0], 2)}"

# 3. Create Gradio Interface
iface = gr.Interface(
    fn=predict_price,
    inputs=[
        gr.Number(label="Year", value=2018),
        gr.Number(label="Mileage", value=50000),
        gr.Number(label="Tax", value=150),
        gr.Number(label="MPG", value=40),
        gr.Number(label="Engine Size", value=2.0)
    ],
    outputs=gr.Textbox(label="Result"),
    title="Ford Car Price Predictor",
    description="Enter car details to predict price"
)

# 4. Launch with public link
iface.launch(share=True)
```
