```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/master/insurance.csv")
print(df.head(10))
```

```python
print("Shape:", df.shape)
print("\nInfo:")
print(df.info())
print("\nDescribe:")
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
print("Numerical Columns:", numerical_cols)
print("Categorical Columns:", categorical_cols)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

numerical_cols = ['age', 'bmi', 'children', 'charges']

for col in numerical_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True, bins=20, color='skyblue') # kde=True साठी distribution curve येईल
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

categorical_cols = ['sex', 'smoker', 'region']
for col in categorical_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(x=df[col])
    plt.title(f'Count Plot of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

numerical_cols = ['age', 'bmi', 'children', 'charges']
plt.figure(figsize=(8,6))
corr = df[numerical_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
print("Average Charges:", df['charges'].mean())
print("Max Charges:", df['charges'].max())
print("Min Charges:", df['charges'].min())
print("\nAverage Charges by Smoker:")
print(df.groupby('smoker')['charges'].mean())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.boxplot(x='smoker', y='charges', data=df)
plt.title('Charges by Smoker')

plt.subplot(1,2,2)
sns.boxplot(x='sex', y='charges', data=df)
plt.title('Charges by Sex')
plt.show()
```

```python
print("1. Average Age:", df['age'].mean())
print("Average BMI:", df['bmi'].mean())
print("\n2. Smoking effect on charges:")
print(df.groupby('smoker')['charges'].mean())
print("\n3. Region with most customers:", df['region'].value_counts().idxmax())
print("\n4. Observation: Charges are much higher for smokers. Age and BMI also increase charges.")
```
