```python
# Cell 1: Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

```python
# Cell 2: Load Dataset + Basic Info
df = pd.read_csv("Titanic.csv")

print("Q1. First 5 records:")
print(df.head())

print("\nQ2. Dataset Shape:", df.shape)

print("\nQ3. Column Names:")
print(df.columns.tolist())

print("\nQ4. Data Types:")
print(df.dtypes)

print("\nQ5. Missing Values:")
print(df.isnull().sum())

print("\nQ6. Summary Statistics:")
print(df.describe())
```

```python
# Cell 4: Visualization 2
plt.figure(figsize=(8,4))
sns.histplot(df['Age'].dropna(), bins=30, kde=True, color='skyblue')
plt.title('Q7.2 Age Distribution')
plt.show()
```

```python
# Cell 5: Visualization 3
plt.figure(figsize=(8,4))
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title('Q7.3 Fare Distribution by Passenger Class')
plt.show()
```

```python
# Cell 6: Visualization 4
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Q7.4 Correlation Heatmap')
plt.show()
```

```python
# Cell 7: Visualization 5
plt.figure(figsize=(6,4))
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Q7.5 Survival Rate by Gender')
plt.show()
```

```python
# Cell 8: Visualization 6
plt.figure(figsize=(6,6))
class_counts = df['Pclass'].value_counts()
plt.pie(class_counts, labels=['Class 1','Class 2','Class 3'], autopct='%1.1f%%')
plt.title('Q7.6 Passenger Class Distribution')
plt.show()
```
